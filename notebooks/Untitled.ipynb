```python
import pandas as pd
```

```python
humans = pd.read_csv('data/responses.csv')
```

```python
humans.head()
```

```python
humans.drop(columns='Timestamp', axis = 'columns', inplace = True)
```

```python
humans_dummies = pd.get_dummies(humans)
humans_dummies.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X = humans_dummies.drop(columns = 'How successful has this tutorial been so far?')
```

```python
y = (humans_dummies['How successful has this tutorial been so far?'] >= 5).astype(int).values
```

```python
y = [1 if x >=5 else 0 for x in humans_dummies['How successful has this tutorial been so far?']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
tree = RandomForestClassifier(max_depth = 10)
```

```python
tree.fit(X_train, y_train)
```

```python
tree.score(X_test, y_test)
```
